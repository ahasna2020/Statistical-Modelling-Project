# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
# read data saved from citybikes API
df_stations = pd.read_csv('C:\\Users\\aysha\\Documents\\Python Scripts\\Statistical-Modelling-Project\\data\\bike_stations.csv', index_col=0)
df_stations.head()

api_key = os.environ["FOURSQUARE_API_KEY"]
# os.environ loads the dictionary with environment variables 
# where os.environ.keys() are all variable names

#df_stations = df_stations.reset_index()  # make sure indexes pair with number of rows

cnt=0
new_df  = pd.DataFrame() 
final_df  = pd.DataFrame() 

#sample url
#url = "https://api.foursquare.com/v3/places/search?radius=1000&ll=42.386781%2C-71.006098"
for index, row in df_stations.iterrows():
    
    url = "https://api.foursquare.com/v3/places/search?radius=1000&ll="+str(row['latitude'])+','+str(row['longitude'])
    headers = {"accept": "application/json"}
    headers['Authorization'] = api_key
    print(url)
    response = requests.get(url, headers=headers)
    data = json.loads(response.text)
    data =data['results']
    
    if cnt==0:
        new_df= pd.json_normalize(data, record_path=['categories'], meta=['fsq_id', 'name', 'distance', ['location', 'address']],meta_prefix='poi_', errors='ignore')
        new_df['free_bikes'] = row['free_bikes']
        new_df['station_lat']=row['longitude']
        new_df['station_long']=row['latitude']
        print('new_df count:',len(new_df))
        final_df = new_df
        cnt+=1
        #print(cnt)
    else:
        new_df= pd.json_normalize(data, record_path=['categories'], meta=['fsq_id', 'name', 'distance', ['location', 'address']],meta_prefix='poi_', errors='ignore')
        new_df['free_bikes'] = row['free_bikes']
        new_df['station_lat']=row['longitude']
        new_df['station_long']=row['latitude']
        final_df=pd.concat([final_df, new_df], 
                  ignore_index = True)
        print('else new_df count:',len(new_df))
        print('else final_df count:',len(final_df))
        cnt+=1
        #print(cnt)
    
    #testing data for a sample dataset
    if cnt==5:
        break
    
print("API parsing completed") 
print("Final dataset row count:")
print(final_df.shape)

#API parsing completed
#Final dataset row count:
#(15, 10)
```

```python
#ACTUAL CODE
# read data saved from citybikes API
df_stations = pd.read_csv('C:\\Users\\aysha\\Documents\\Python Scripts\\Statistical-Modelling-Project\\data\\bike_stations.csv', index_col=0)
df_stations.head()

api_key = os.environ["YELP_API_TOKEN"]
# os.environ loads the dictionary with environment variables 
# where os.environ.keys() are all variable names


cnt=0
new_df  = pd.DataFrame() 
final_df  = pd.DataFrame() 

#sample url
#url = "https://api.yelp.com/v3/businesses/search?latitude=42.386781&longitude=-71.006098&radius=1000&sort_by=best_match"
for index, row in df_stations.iterrows():
    
    url = "https://api.yelp.com/v3/businesses/search?radius=1000&sort_by=best_match&latitude="+str(row['latitude'])+'&longitude='+str(row['longitude'])
    headers = {"accept": "application/json"}
    headers['Authorization'] = api_key
    print(url)
    
    #call API
    response = requests.get(url, headers=headers)
    data = json.loads(response.text)
    data =data['businesses']
    
    if cnt==0:
        new_df=final_df = pd.json_normalize(data, record_path=['categories'], meta=['id', 'name', 'distance', ['location', 'address1'], 'rating'],meta_prefix='poi_', errors='ignore')
        new_df['free_bikes'] = row['free_bikes']
        new_df['station_lat']=row['longitude']
        new_df['station_long']=row['latitude']
        final_df=new_df
        print('new_df count:',len(new_df))
        cnt+=1
        #print(cnt)
    else:
        new_df= pd.json_normalize(data, record_path=['categories'], meta=['id', 'name', 'distance', ['location', 'address1'], 'rating'],meta_prefix='poi_', errors='ignore')
        new_df['free_bikes'] = row['free_bikes']
        new_df['station_lat']=row['longitude']
        new_df['station_long']=row['latitude']
        final_df=pd.concat([final_df, new_df], ignore_index = True)
        print('else new_df count:',len(new_df))
        print('else final_df count:',len(final_df))
        cnt+=1
        #print(cnt)
    
    #testing data for a sample dataset
    if cnt==5:
        break
    
print("API parsing completed") 
print("Final dataset row count:")
print(final_df.shape)

#API parsing completed
#Final dataset row count:
#(41, 7)
```

```python
#get the restaurants and their rating
df_poi = pd.read_csv('C:\\Users\\aysha\\Documents\\Python Scripts\\Statistical-Modelling-Project\\data\\Yelp_data.csv', index_col=0)
df_poi.head(40)

#get restaurants
mask = df_poi['poi_name'].str.contains('Restaurant', case=False, na=False)

restaurants = df_poi[mask]

restaurants = restaurants[['poi_name','poi_rating']]

restaurants_df = pd.DataFrame(restaurants) 

#remove duplicates
restaurants_df.drop_duplicates(inplace=True)

#restaurants_df = restaurants_df.reset_index()

restaurants_df.sort_values('poi_rating',inplace=True, ascending=False)

restaurants_df.head(10)

```
