# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import os


url = "https://api.citybik.es/v2/networks/blue-bikes"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
import json

my_json= response.text

#load the json data
data = json.loads(my_json)

#get the stations data
stations = data['network']['stations']

# Create an empty dictionary
stations_dict = [{}]
additional_data={}


cnt=0
for station in stations:
    # Define a dictionary with key-value pairs to add
    additional_data = {'latitude': station['latitude'], 'longitude': station['longitude'], 'free_bikes': station['free_bikes']}
    if len(additional_data)>0:
        stations_dict.append(additional_data)
        
```

Put your parsed results into a DataFrame.

```python
import pandas as pd

df=pd.DataFrame(data=stations_dict,columns=['latitude','longitude','free_bikes'])
```

```python
#drop any null rows
df= df.dropna()

print('shape:', df.shape)
```
